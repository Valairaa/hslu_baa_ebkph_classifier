# Data visualization and analysis

## 1. Loading and preparing data

```python
# imports
import pandas as pd
import sys

# helpers
sys.path.insert(0, "../../")
from geometric_extraction_helper import GENERAL_KEYS, AABB_KEYS, TFBB_KEYS, TOPO_KEYS, RAY_KEYS, MATERIAL_KEYS, ALL_FEATURE_KEYS
```

```python
# data directory and output path
DATA_MODELS_DIR = "../3_4_extracting geometric features/data_geometric_featueres_enriched.parquet"
OUTPUT_PATH = "./data_dropped_input_duplicates.parquet"
KAEGGLE_OUTPUT_PATH = "./data_eBKPh_kaggle.csv"

# load parquet file
df = pd.read_parquet(DATA_MODELS_DIR)
df.head()
```

## 2. Analyse duplicates in the input features

```python
# show duplicates of general geometric features
df[GENERAL_KEYS].duplicated(keep=False).sum()
```

```python
# show duplicates of aabb features
df[AABB_KEYS].duplicated(keep=False).sum()
```

```python
# show duplicates of tfbb features
df[TFBB_KEYS].duplicated(keep=False).sum()
```

```python
# show duplicates of topological features
df[TOPO_KEYS].duplicated(keep=False).sum()
```

```python
label_cols = [c for c in df.columns if c.startswith("label_")]
label_cols
```

```python
# show duplicates of all input features and labels
# duplicates are only elements with identical features and identical labels
duplicates = df[ALL_FEATURE_KEYS + label_cols].duplicated(keep=False)
print(f"Count of duplicated input features: {duplicates.sum()} ({duplicates.sum() / len(df) * 100:.1f}%)")

# look at the duplicated rows to check them manually
df[duplicates].sort_values(ALL_FEATURE_KEYS).head(10)
```

```python
# drop duplicates, which are identical on all four input categories -> means identical objects
# we can drop them without losing information
df = df[~duplicates]
```

```python
# check single categories again to see how many duplicates remain
df[GENERAL_KEYS].duplicated(keep=False).sum(), df[AABB_KEYS].duplicated(keep=False).sum(), df[TFBB_KEYS].duplicated(keep=False).sum(), df[TOPO_KEYS].duplicated(keep=False).sum(), df[MATERIAL_KEYS].duplicated(keep=False).sum(), df[RAY_KEYS].duplicated(keep=False).sum()
```

```python
# remaining size of the dataset after dropping duplicates
len(df)
```

```python
# show nan values of all input features
na_counts = df[ALL_FEATURE_KEYS].isna().sum()
na_counts[na_counts > 0]
```

```python
# drop nan values with label geom_cols -> geometric extraction failed, so we cannot use these samples for training
df = df.dropna(subset=GENERAL_KEYS + AABB_KEYS + TFBB_KEYS + TOPO_KEYS)
print(df[GENERAL_KEYS + AABB_KEYS + TFBB_KEYS + TOPO_KEYS].isna().sum())
```

```python
# replace NaN values with -1 for materials and raycasting features
# model needs to learn which elements to not have materials
# model neefs to learn which elements are not horizontal elements
df[RAY_KEYS] = df[RAY_KEYS].fillna(-1)
df[MATERIAL_KEYS] = df[MATERIAL_KEYS].fillna(-1)
```

```python
# show nan values of all input features again
na_counts = df[ALL_FEATURE_KEYS].isna().sum()
na_counts[na_counts > 0]
```

## 3. Analyse "IfcBuildingElementProxy" class and its subclasses

## 4. Analyse outliers in the input features

```python
# get summary of outliers in the input features
# leave it at the moment, because they must be removed for KNN or SVM, but not for tree-based models like Random Forest or XGBoost
outlier_summary = []
for col in ALL_FEATURE_KEYS:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    n_outliers = ((df[col] < Q1 - 1.5 * IQR) | (df[col] > Q3 + 1.5 * IQR)).sum()
    outlier_summary.append({
        "feature": col,
        "n_outliers": n_outliers,
        "pct": round(n_outliers / len(df) * 100, 2),
        "min": df[col].min(),
        "max": df[col].max(),
        "Q1": Q1,
        "Q3": Q3,
    })

# show most 60 features with the highest percentage of outliers
pd.DataFrame(outlier_summary).set_index("feature").sort_values("pct", ascending=False).head(n=60)
```

## 4. Save the cleaned dataset

```python
# reset index after dropping rows with NaN values
df = df.reset_index(drop=True)

# save cleaned dataframe to parquet file
df.to_parquet(OUTPUT_PATH, index=False)
```

## 5. Save dataset for kaeggle with anonymized project codes

```python
# map each project code to project_1, project_2, ...
unique_projects = df["project_code"].unique()
project_mapping = {code: f"project_{i + 1}" for i, code in enumerate(unique_projects)}

# anonymized project code + guid + ifc_schema + the 73 extracted features
kaeggle_df = pd.DataFrame({
    "project": df["project_code"].map(project_mapping),
    "guid": df["guid"],
    "ifc_schema": df["ifc_schema"],
})
kaeggle_df = pd.concat([kaeggle_df, df[ALL_FEATURE_KEYS]], axis=1)

# save anonymized dataset to csv
kaeggle_df.to_csv(KAEGGLE_OUTPUT_PATH, index=False)
print(f"Saved {len(kaeggle_df)} rows x {kaeggle_df.shape[1]} columns to {KAEGGLE_OUTPUT_PATH}")
kaeggle_df.head()
```
