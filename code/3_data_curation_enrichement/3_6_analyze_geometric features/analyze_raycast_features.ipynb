# Data visualization and analysis

## 1. Loading and preparing data

```python
# imports
from sklearn.preprocessing import PowerTransformer
import pandas as pd
import sys

# helpers
sys.path.insert(0, "../../")
from dataviz_helper import (
    plot_label_property_status,
    plot_float_boxplots
)

from geometric_extraction_helper import RAY_KEYS

# define PowerTransformer for skewed features
# important for KNN or SVM, irrelevant for tree-based models like Random Forest or XGBoost
pt = PowerTransformer(method="yeo-johnson")
```

```python
# data directory and output path
DATA_MODELS_DIR = "../3_5_clean_all_features/data_dropped_input_duplicates.parquet"

# load parquet file
df = pd.read_parquet(DATA_MODELS_DIR)
df.head()
```

## 2. Analyse the distribution of raycasting features

```python
# show distribution of raycasting properties
plot_label_property_status(df, cols=RAY_KEYS, title="Verteilung Features (Anzahl Elemente ober- oder unterhalb horizontaler Elemente)")
```

```python
# shows boxplots of rayscasting features before deskewing
plot_float_boxplots(df, RAY_KEYS, title="Boxplots: Features vor der Transformation (Anzahl Elemente ober- oder unterhalb horizontaler Elemente)")
```

```python
# clean skewed features with power transformation
df[RAY_KEYS] = pt.fit_transform(df[RAY_KEYS])
```

```python
# shows boxplots of rayscasting features after deskewing
plot_float_boxplots(df, RAY_KEYS, title="Boxplots: Features nach der Transformation (Anzahl Elemente ober- oder unterhalb horizontaler Elemente)")
```

## Conclusion

Many different input geometric features are skewed and have outliers. Some ML models are sensitive to skewed features and outliers, so it is important to handle them before training the models. In the other hand, models like Random Forest are not sensitive to skewed features and outliers, so it is not necessary to handle them for these models. Because of this, the data will be deskwed in the transformation Pipeline for each model separately like SVM or KNN.
