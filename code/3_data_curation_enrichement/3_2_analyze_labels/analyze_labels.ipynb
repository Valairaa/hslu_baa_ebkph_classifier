## 3. Rename minor labels to minor labels in the dataset

```python
minor_labels = ["label_aussenliegendes_bauteil",
                "label_konstruktionsergaenzung",
                "label_unter_terrain",
                "label_deckbelag",
                "label_bekleidung",
                "label_erdverbunden",
                "label_unterkonstruktion",
                "label_verdunkelung",
                "label_schutzschicht",
                "label_sonnenschutz",
                "label_einbau",
                "label_aufzugstyp"
                ]
```

```python
# rename columns in the dataset to minor_label
for col in minor_labels:
    df.rename(columns={col: col.replace("label_", "minor_label_")}, inplace=True)
df.columns
```

```python
# show all label columns in the dataset
label_cols = [c for c in df.columns if c.startswith("label_")]
label_cols
```

```python
# show all minor labels in the dataset
minor_label_cols = [c for c in df.columns if c.startswith("minor_label_")]
minor_label_cols
```

```python
# save as parquet
output_path = Path("data_eBKPh_labels_renamed.parquet")
df.to_parquet(output_path, index=False)
```
