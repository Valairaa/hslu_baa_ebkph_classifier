# Collecting label data from excel files

```python
# imports
from pathlib import Path
import pandas as pd
import sys

# helpers
sys.path.insert(0, "../../")
from dataloader import extract_excel_label_informations
```

```python
# data directory and output path
EXCEL_MODELS_DIR = "../../0_data/1_ifc_models"
OUTPUT_PATH = "label_data.parquet"
```

```python
# get locations of ifc files
xlsx_files = sorted(Path(EXCEL_MODELS_DIR).rglob("*.xlsx"), reverse=True)

# print count and example
print("Count of Excel-Files: ", len(xlsx_files))
xlsx_files[0]
```

```python
# filter out not necessary excel file, only label files are needed
label_files = [f for f in xlsx_files if "labels" in f.name.lower()]
len(label_files)
```

```python
label_files
```

```python
# process each ifc file and save data to .parquet
all_records = []
for excel_path in label_files:
    print(f"  Processing: {excel_path.name}")
    all_records.extend(extract_excel_label_informations(excel_path))

df = pd.DataFrame(all_records)
df
```

```python
# merge Modell and Kurzname to one column
df["Modell"] = df["Modell"].fillna("") + " " + df["Kurzname"].fillna("")
df = df.drop(columns=["Kurzname"])
df
```

```python
# remove column Modell
df = df.drop(columns=["Farbe"])
df
```

```python
# rename columns
df = df.rename(columns={
    "Modell": "model_name",
    "Geschoss": "storey",
    "GKS Typ": "type_gks",
    "IFC Entität": "ifc_entity",
    "Predefined Type (Vordefinierter Typ)": "predefined_type",
    "GUID": "guid",
    "CHE_eBKP-H 2020 GKS": "eBKPh"
})
df
```

```python
# show project code distribution
df["model_name"].value_counts()
```

```python
# save labels
df.to_parquet(OUTPUT_PATH, index=False)

print(f"\nSaved {len(df):,} rows to: {OUTPUT_PATH}\n")
```
