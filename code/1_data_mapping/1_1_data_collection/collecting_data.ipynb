# Collecting base data from ifc files

```python
# imports
from pathlib import Path
import pandas as pd
import sys

# helpers
sys.path.insert(0, "../../")
from dataloader import extract_ifc_main_properties
```

```python
# data directory and output path
IFC_MODELS_DIR = "../../0_data/1_ifc_models"
OUTPUT_PATH = "ifc_data.parquet"
```

```python
# get locations of ifc files
ifc_files = sorted(Path(IFC_MODELS_DIR).rglob("*.ifc"), reverse=True)

# print count and example
print("Count of IFC-Files: ", len(ifc_files))
ifc_files[0]
```

```python
# process each ifc file and save data to .parquet
all_records = []
for ifc_path in ifc_files:
    print(f"  Processing: {ifc_path.name}")
    all_records.extend(extract_ifc_main_properties(ifc_path))

df = pd.DataFrame(all_records)
df
```

```python
# save ifc base data
df.to_parquet(OUTPUT_PATH, index=False)

print(f"\nSaved {len(df):,} rows to: {OUTPUT_PATH}\n")
```
