# Map collected raw and labeled data

```python
# imports
import pandas as pd
```

```python
# data directory and output path
DATA_DIR = "../../1_data_mapping/1_1_data_collection/ifc_data.parquet"
LABEL_DIR = "../../1_data_mapping/1_1_data_collection/label_data.parquet"
OUTPUT_PATH = "mapped_data.parquet"
```

```python
# load data
ifc_df = pd.read_parquet(DATA_DIR)
label_df = pd.read_parquet(LABEL_DIR)
```

```python
# deduplicate data
print("Size of label data before deduplication:", len(label_df))
label_df = label_df.drop_duplicates(["guid", "ifc_entity"])
print("Size of label data after deduplication:", len(label_df))
```

```python
# merge on common keys
df = ifc_df.merge(
    label_df,
    on=["guid", "ifc_entity"],
    how="left",
)
```

```python
# extract project code from model name, rename and drop columns
df["project_code"] = df["model_name_x"].str.split("_").str[0]
df = df.drop(columns=["model_name_y"])
df = df.rename(columns={"model_name_x": "model_name"})
df = df.rename(columns={"file_path": "ifc_file_path"})
```

```python
# save
df.to_parquet(OUTPUT_PATH, index=False)
print(f"IFC rows: {len(ifc_df):}")
print(f"Label rows: {len(label_df):,}")
print(f"Merged rows: {len(df):,}")
```
