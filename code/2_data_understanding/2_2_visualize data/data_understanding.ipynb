## 4.1 Calculate co-label fingerprint

```python
# boolean columns to long format
code_cols = [c for c in df.columns if c.startswith('eBKPh_') and c not in ('eBKPh_original', 'eBKPh_status') and df[c].dtype == bool]

df_melted = df[["guid"] + code_cols].melt(id_vars="guid", var_name="ebkp_code", value_name="has_label")
df_melted = df_melted[df_melted["has_label"]]
df_melted["ebkp_code"] = df_melted["ebkp_code"].str.replace("eBKPh_", "").str.replace("_", ".")

# group all codes per element – once per level
def top_combos_for_level(df_melted, pattern, n=30):
    filtered = df_melted[df_melted["ebkp_code"].str.match(pattern)]
    label_sets = filtered.groupby("guid")["ebkp_code"].apply(frozenset).reset_index()
    combo_counts = Counter(label_sets["ebkp_code"])
    return sorted(combo_counts.items(), key=lambda x: -x[1])[:n]
```

```python
# show top combinations of eBKPh labels at letter level
print("Letter levels (z.B. C, D, E)\n")
for combo, count in top_combos_for_level(df_melted, r'^[A-Z]$'):
    print(f"{count:>6}x  {sorted(combo)}")
```

```python
# show top combinations of eBKPh labels at two-digit level
print("two-digit level (z.B. C01, D04)\n")
for combo, count in top_combos_for_level(df_melted, r'^[A-Z]\d{2}$'):
    print(f"{count:>6}x  {sorted(combo)}")
```

```python
# show top combinations of eBKPh labels at four-digit level
print("four-digit level (z.B. C01.01, D04.01)\n")
for combo, count in top_combos_for_level(df_melted, r'^[A-Z]\d{2}\.\d{2}$'):
    print(f"{count:>6}x  {sorted(combo)}")
```
