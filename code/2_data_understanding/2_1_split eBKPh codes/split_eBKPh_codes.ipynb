## 5. Split eBKPh code into own columns

```python
# show all values
df["eBKPh"].value_counts()
```

```python
# define regex pattern and method for eBKPh codes
CODE_PATTERN = re.compile(r'\b[A-Z]\d{2}\.\d{2}|\b[A-Z]\d{2}|\b[A-Z](?=\s|$)')

# method to extract eBKPh codes from string, returns list of codes
def parse_eBKPh_codes(val):
    if pd.isna(val):
        return []
    return CODE_PATTERN.findall(str(val))
```

```python
# apply method to create new columns for each code
df["eBKPh_original"] = df["eBKPh"]
df["eBKPh_parsed"] = df["eBKPh"].apply(parse_eBKPh_codes)
all_codes = sorted(set(code for codes in df["eBKPh_parsed"] for code in codes))

# create new columns for each code and set to True if code is present in the parsed list
for code in all_codes:
    col_name = f"eBKPh_{code.replace('.', '_')}"  # z.B. "ebkp_C2_1"
    df[col_name] = df["eBKPh_parsed"].apply(lambda codes: code in codes)

# drop the temporary parsed column
df.drop(columns=["eBKPh_parsed"], inplace=True)

print(df.head())
print(f"\nCreated columns: {[c for c in df.columns if c.startswith('eBKPh_') and c != 'eBKPh_original']}")
```

## 6. Enrich the main codes from their subcodes

```python
# propagate subcodes to parent codes for a hierarchical consistency
for col in [c for c in df.columns if c.startswith('eBKPh_') and c not in ('eBKPh_original', 'eBKPh_status')]:
    code = col.replace('eBKPh_', '').replace('_', '.')

    if re.match(r'^[A-Z]\d{2}\.\d{2}$', code):           # four_digit: C02.01
        parent_two    = f"eBKPh_{code[:3]}"              # -> C02
        parent_letter = f"eBKPh_{code[0]}"               # -> C
        if parent_two in df.columns:
            df[parent_two] = df[parent_two] | df[col]
        if parent_letter in df.columns:
            df[parent_letter] = df[parent_letter] | df[col]

    elif re.match(r'^[A-Z]\d{2}$', code):                # two_digit: C02
        parent_letter = f"eBKPh_{code[0]}"               # -> C
        if parent_letter in df.columns:
            df[parent_letter] = df[parent_letter] | df[col]
```

```python
# save as parquet
df.to_parquet(OUTPUT_PATH, index=False)
print(f"DataFrame with {len(df)} rows and {len(df.columns)} columns were saved as: '{OUTPUT_PATH}'")
```
